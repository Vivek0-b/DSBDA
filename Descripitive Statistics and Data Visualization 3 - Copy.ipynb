```python
def mean(data):
    Sum = 0
    for i in data:
        Sum += i
    return Sum / len(data)

def minimum(data):
    if len(data) == 0:
        return None
    smallest = data[0]
    for number in data[1:]:
        if number < smallest:
            smallest = number
    return smallest

def maximum(data):
    if len(data) == 0:
        return None
    largest = data[0]
    for number in data[1:]:
        if number > largest:
            largest = number
    return largest

def median(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    middle = n // 2
    if n % 2 == 0:
        return (sorted_data[middle - 1] + sorted_data[middle]) / 2
    else:
        return sorted_data[middle]

def count(data):
    return len(data)

def std(data):
    mean_value = mean(data)
    variance = sum((x - mean_value) ** 2 for x in data) / len(data)
    return variance ** 0.5

def quantile(data, q):
    sorted_data = sorted(data)
    index = int(q * len(sorted_data))
    return sorted_data[index]

def dspstat(data, data_name):
    numeric_columns = ['age']
    print(f"\nStatistics ({data_name})")
    for column in numeric_columns:
        column_data = data[column].values
        print(f"\nStatistics for {column}:")
        print("Mean:", mean(column_data))
        print("Minimum:", minimum(column_data))
        print("Maximum:", maximum(column_data))
        print("Median:", median(column_data))
        print("Count:", count(column_data))
        print("Standard deviation:", std(column_data))
        print("25th Percentile:", quantile(column_data, 0.25))
        print("75th Percentile:", quantile(column_data, 0.75))

age_data = data[['age']]
dspstat(age_data, 'Age Data')
```
